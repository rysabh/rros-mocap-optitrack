```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
trials = [90] # list(range(1,27))

for trial in trials: 
    dir = f"~/exp_exec_data/trial_{trial}/csv/"
    CPE_file = dir + "CPE.csv" 
    FPE_file = dir + "FPE.csv"  
    CPE = pd.read_csv(CPE_file)  
    FPE = pd.read_csv(FPE_file)  

    save_dir = "/home/aero/exp_exec_data/plots"

    dims_plot = ['X','Y','A','B','C'] 
    dims_range = [10,10,10,10,10]
    plot_coord = [[0,0],[0,1],[1,0],[1,1],[1,2]] 

    fig,axs = plt.subplots(2,3,figsize=(20,10))
    fig.suptitle(f"Trial {trial}")  
    for i, dim in enumerate(dims_plot): 
        axs_handle = axs[plot_coord[i][0],plot_coord[i][1]] 
        axs_handle.hist(CPE[dim+"_err"], bins='auto') 
        axs_handle.axvline(FPE[dim+"_err"][1],color='red')    
        axs_handle.axvline(0,color='green')    
        axs_handle.set_xlim(-dims_range[i], +dims_range[i]) 
        axs_handle.set_title(dim+" Error")  
        
    plt.savefig(save_dir+f"/Trial_{trial}.png")  
```
